```python
V = {}
keys = ['alpha', 'beta', 'gamma', 'delta', 'R0', 'F0', 'T', 'dt']
for i, key in enumerate(keys):
    V[key] = float(input('Please provide the value of ' + key +' : '))
```

```python
def odes(u, t):
    R,F = u
    first = V['alpha']*R - V['beta']*R*F
    second = -V['gamma']*F + V['delta']*R*F
    return first, second
```

```python
u0 = np.array([V['R0'], V['F0']])
```

```python
from scipy.integrate import odeint
t = np.arange(0,V['T'], V['dt'])
u = odeint(odes, u0, t)
```

```python
R = u[:,0]
F = u[:,1]

dRdt = V['alpha']*R - V['beta']*R*F
dFdt = -V['gamma']*F + V['delta']*R*F
```

```python
plt.figure(figsize = (12,7))

plt.subplot(2,4,1)
plt.plot(t,R)
plt.xlabel('t')
plt.ylabel('R(t)')

plt.subplot(2,4,2)
plt.plot(t,F)
plt.xlabel('t')
plt.ylabel('F(t)')

plt.subplot(2,4,3)
plt.plot(t,dRdt)
plt.xlabel('t')
plt.ylabel('dRdt(t)')

plt.subplot(2,4,4)
plt.plot(t,dFdt)
plt.xlabel('t')
plt.ylabel('dFdt(t)')

plt.subplot(2,4,5)
plt.plot(R,dRdt)
plt.xlabel('R')
plt.ylabel('dRdt(t)')

plt.subplot(2,4,6)
plt.plot(F,dFdt)
plt.xlabel('F')
plt.ylabel('dFdt(t)')

plt.subplot(2,4,7)
plt.plot(F,R)
plt.xlabel('F(t)')
plt.ylabel('R(t)')

plt.tight_layout()
plt.show()
```

```python
I = V['delta']*R - V['gamma']*np.log(R) + V['beta']*F - V['alpha']*np.log(F)
plt.plot(t,I);
```
