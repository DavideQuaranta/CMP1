```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81
m = 1
L = 1
theta = np.radians(0.1)
v0 = 0
beta = 0
T = 2
dt = 0.001
```

```python
x = np.array([L*np.sin(theta)])
y = np.array([L*np.cos(theta)])
vx = np.array([-v0*np.cos(theta)])
vy = np.array([-v0*np.sin(theta)])
t = np.array([0])
```

```python
for i in range(int(T/dt)):
    vx = np.append(vx, vx[i] + dt*(-g*x[i]/L - beta*vx[i]/m))
    vy = np.append(vx, vy[i] + dt*(-g + g*y[i]/L - beta*vy[i]/m))
    x = np.append(x, x[i] + vx[i]*dt)
    y = np.append(y, y[i] + vy[i]*dt)
    t = np.append(t, i*dt)
```

```python
%matplotlib inline
plt.scatter(x,y)
```

```python

t = t[0::50]
x = x[0::50]
```

```python
y = y[0::50]
```

```python
import matplotlib.animation as animation
%matplotlib notebook
# Create a figure object
fig = plt.figure()

ax = fig.add_subplot(111, autoscale_on=True, xlim=(min(x)*1.2, max(x)*1.2), ylim=(min(y)*1.2,max(y)*1.2))
ax.grid()
ax.set_xlabel('x(t) [m]')
ax.set_ylabel("y(t) [m]")
# plt.title("Trajectory of a projectile with $v_0$: %.1f m/s\t $\Theta_0$: %.1f$^\circ$" % (v0,angle))


line1 = ax.plot(x[0], y[0], '.', lw=2)

def update_plots(i):
    
    line1[0].set_data(x[:i], y[:i])   
    return line1,

anim = animation.FuncAnimation(fig, update_plots, np.arange(1, len(x)), interval=0.5, blit=True, repeat=True)

plt.show()
```
